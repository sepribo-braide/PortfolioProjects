```python

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

#Read in the data
df = pd.read_csv(r"C:\Users\SEPRIBO\Desktop\Data Analyst Portfolio\movies.csv")
```

```python
#Let's look at the data

df.head()
```

```python
#Let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
#data types per columns
df.dtypes
```

```python
#Change data type of columns
df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
df.head()
```

```python
df.head()
```

```python

#Create correct year column
df['yearcorrect'] = df['released'].astype(str).str.split().str[2]
df.head()
```

```python
#Order the data by gross revenue
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
#Checking for duplicates, drop any duplicates
df['company'].drop_duplicates().sort_values(ascending=False).head()
```

```python
#Budget will have high correlation
#Company hig correlation
```

```python
#Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.ylabel('Gross Earnings')
plt.xlabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
#Plot the budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data = df, scatter_kws={"color": "red"}, line_kws={"color": "green"})

plt.show()
```

```python
#Lets start looking at correlation

# 1. Select only numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])

# 2. Calculate correlation on numeric columns only
numeric_df.corr(method='pearson') #pearson, kendall, spearman
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix=numeric_df.corr(method='spearman')

sns.heatmap(correlation_matrix, annot=True)


plt.title('Correlation Matrix for Numeric Features')
plt.ylabel('Movie Features')
plt.xlabel('Movie Features')



plt.show()
```

```python
#look at company
df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
correlation_matrix=df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)


plt.title('Correlation Matrix for Numeric Features')
plt.ylabel('Movie Features')
plt.xlabel('Movie Features')



plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
